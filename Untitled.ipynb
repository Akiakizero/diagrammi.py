```python
import random
def failsafe():
    input = ""
    print("Please type 'failsafe' and press enter to start playing!")
    while input != "failsafe":
        input = input()

def makePipe(height):
    outL = [""] * (12)

    pipeGapHeight = 2
    for i in range(0, 11 + 1, 1):
        if fabs(i - height) < pipeGapHeight:
            outL[i] = " "
        else:
            if fabs(i - height) < pipeGapHeight + 1 and fabs(i - height) == i - height:
                outL[i] = "┳"
            else:
                if fabs(i - height) < pipeGapHeight + 1 and fabs(i - height) != i - height:
                    outL[i] = "ꓕ"
                else:
                    outL[i] = "|"
    out = render(outL, False)
    
    return out

def makeScreen(y, world):
    allCols = [""] * (len(world))

    for i in range(0, len(world) - 1 + 1, 1):
        if world[i] == 0:
            allCols[i] = ""
            for j in range(0, 22 + 1, 1):
                allCols[i] = allCols[i] + " "
        else:
            allCols[i] = makePipe(world[i])
    out = ""
    for i in range(0, len(allCols[0]) - 1 + 1, 1):
        if i == y * 2:
            out = out + "B"
        else:
            out = out + allCols[0][i]
    allCols[0] = out
    screen = render(allCols, True)
    
    return screen

def render(arr, isRow):
    out = ""
    if isRow:
        length = 0
        for i in range(0, len(arr) - 1 + 1, 1):
            if len(arr[i]) > length:
                length = len(arr[i])
        for i in range(0, length - 1 + 1, 1):
            anyInRow = False
            for j in range(0, len(arr) - 1 + 1, 1):
                if len(arr[j]) > i * 2:
                    out = out + arr[j][i * 2]
                    anyInRow = True
            if anyInRow:
                out = out + chr(13)
    else:
        for i in range(0, len(arr) - 1 + 1, 1):
            out = out + arr[i] + chr(13)
    
    return out

# Main
random.seed()   #Prepare random number generator

failsafe()
y = 5
score = 0
loose = False
world = [0] * (20)

for i in range(0, len(world) - 1 + 1, 1):
    world[i] = 0
gap = 0
last = int(random.random() * 11) + 1
while not loose:
    out = input()
    if out != "":
        y = y - 1
    else:
        y = y + 0.5
    if y >= 11:
        y = 11
    else:
        if y <= 0:
            y = 0
    if world[0] != 0:
        if fabs(int(y) - world[0]) > 2:
            loose = True
        else:
            score = score + 1
    for i in range(0, len(world) - 1 + 1, 1):
        if i == len(world) - 1:
            if gap == 0:
                last = last + int(random.random() * 5) - 2
                if last > 12:
                    last = 12
                else:
                    if last < 0:
                        last = 0
                world[i] = last
                gap = 8
            else:
                gap = gap - 1
                world[i] = 0
        else:
            world[i] = world[i + 1]
    if loose:
        pass
    else:
        print(makeScreen(int(y), world) + chr(13) + "Score: " + str(score))
print("You loose! Better luck next time. Remember to press the Q key to quit the python program too!")
print("Score: " + str(score))
```
